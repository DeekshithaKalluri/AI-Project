```python
def create_grid_graph(rows, cols, obstacle_prob=0.2):
    G = nx.grid_2d_graph(rows, cols)
    obstacles = set()

    for node in G.nodes:
        if random.random() < obstacle_prob:
            obstacles.add(node)

    # Remove obstacle nodes from the graph
    G.remove_nodes_from(obstacles)

    return G, obstacles
```

```python
def draw_grid(G, obstacles, path=None):
    pos = {node: (node[1], -node[0]) for node in G.nodes}
    nx.draw(G, pos, node_size=50, node_color='lightgray', with_labels=False)

    if path:
        nx.draw_networkx_nodes(G, pos, nodelist=path, node_color='red', node_size=80)

    if obstacles:
        obstacle_pos = {node: (node[1], -node[0]) for node in obstacles}
        nx.draw_networkx_nodes(G, obstacle_pos, nodelist=obstacles, node_color='black', node_size=60)

# Example usage
rows, cols = 10, 10
G, obstacles = create_grid_graph(rows, cols)
draw_grid(G, obstacles)
```

```python
def manhattan_distance(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

def astar_search(G, start, goal):
    open_set = []
    heapq.heappush(open_set, (0 + manhattan_distance(start, goal), 0, start))
    came_from = {}
    g_score = {start: 0}

    while open_set:
        _, current_g, current = heapq.heappop(open_set)

        if current == goal:
            # Reconstruct path
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)
            return path[::-1]

        for neighbor in G.neighbors(current):
            tentative_g = current_g + 1  # cost between neighbors is 1
            if neighbor not in g_score or tentative_g < g_score[neighbor]:
                g_score[neighbor] = tentative_g
                f_score = tentative_g + manhattan_distance(neighbor, goal)
                heapq.heappush(open_set, (f_score, tentative_g, neighbor))
                came_from[neighbor] = current

    return None  # No path found
```

```python
# Pick start and goal nodes that are not in obstacles
start = (0, 0)
goal = (9, 9)

if start in G.nodes and goal in G.nodes:
    path = astar_search(G, start, goal)
    if path:
        print("Path found!")
        draw_grid(G, obstacles, path)
    else:
        print("No path found.")
else:
    print("Start or goal is inside an obstacle. Try different nodes.")
```

```python
def benchmark_astar(G, start, goal):
    if start not in G.nodes or goal not in G.nodes:
        return None, "Start or goal is invalid."

    start_time = time.perf_counter()
    path = astar_search(G, start, goal)
    end_time = time.perf_counter()

    if path:
        path_length = len(path) - 1  # number of steps, not nodes
        run_time = end_time - start_time
        return {
            "Path Length": path_length,
            "Execution Time (s)": run_time,
            "Path": path
        }, None
    else:
        return None, "No path found."

# Run benchmark
result, error = benchmark_astar(G, start, goal)

if result:
    print("✅ Benchmark Results:")
    for k, v in result.items():
        if k != "Path":
            print(f"{k}: {v}")
else:
    print(f"⚠️ {error}")
```

```python
def dijkstra_search(G, start, goal):
    open_set = []
    heapq.heappush(open_set, (0, start))
    came_from = {}
    g_score = {start: 0}

    while open_set:
        current_g, current = heapq.heappop(open_set)

        if current == goal:
            # Reconstruct path
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)
            return path[::-1]

        for neighbor in G.neighbors(current):
            tentative_g = current_g + 1
            if neighbor not in g_score or tentative_g < g_score[neighbor]:
                g_score[neighbor] = tentative_g
                heapq.heappush(open_set, (tentative_g, neighbor))
                came_from[neighbor] = current

    return None
```

```python
def benchmark_dijkstra(G, start, goal):
    if start not in G.nodes or goal not in G.nodes:
        return None, "Start or goal is invalid."

    start_time = time.perf_counter()
    path = dijkstra_search(G, start, goal)
    end_time = time.perf_counter()

    if path:
        path_length = len(path) - 1
        run_time = end_time - start_time
        return {
            "Path Length": path_length,
            "Execution Time (s)": run_time,
            "Path": path
        }, None
    else:
        return None, "No path found."

# Run benchmark
result_d, error_d = benchmark_dijkstra(G, start, goal)

if result_d:
    print("✅ Dijkstra Benchmark Results:")
    for k, v in result_d.items():
        if k != "Path":
            print(f"{k}: {v}")
else:
    print(f"⚠️ {error_d}")
```

```python
def dfs_limited(G, current, goal, depth, path, visited):
    if depth == 0 and current == goal:
        return path + [current]
    if depth > 0:
        visited.add(current)
        for neighbor in G.neighbors(current):
            if neighbor not in visited:
                result = dfs_limited(G, neighbor, goal, depth - 1, path + [current], visited)
                if result:
                    return result
    return None

def iddfs_search(G, start, goal, max_depth=50):
    for depth in range(max_depth):
        visited = set()
        result = dfs_limited(G, start, goal, depth, [], visited)
        if result:
            return result
    return None
```

```python
def benchmark_iddfs(G, start, goal):
    if start not in G.nodes or goal not in G.nodes:
        return None, "Start or goal is invalid."

    start_time = time.perf_counter()
    path = iddfs_search(G, start, goal)
    end_time = time.perf_counter()

    if path:
        path_length = len(path) - 1
        run_time = end_time - start_time
        return {
            "Path Length": path_length,
            "Execution Time (s)": run_time,
            "Path": path
        }, None
    else:
        return None, "No path found."

# Run benchmark
result_iddfs, error_iddfs = benchmark_iddfs(G, start, goal)

if result_iddfs:
    print("✅ IDDFS Benchmark Results:")
    for k, v in result_iddfs.items():
        if k != "Path":
            print(f"{k}: {v}")
else:
    print(f"⚠️ {error_iddfs}")
```

```python
def beam_search(G, start, goal, beam_width=3):
    from queue import PriorityQueue

    queue = PriorityQueue()
    queue.put((manhattan_distance(start, goal), [start]))

    while not queue.empty():
        new_queue = PriorityQueue()
        width = 0

        while not queue.empty() and width < beam_width:
            _, path = queue.get()
            current = path[-1]

            if current == goal:
                return path

            for neighbor in G.neighbors(current):
                if neighbor not in path:  # avoid cycles
                    new_path = path + [neighbor]
                    priority = manhattan_distance(neighbor, goal)
                    new_queue.put((priority, new_path))
                    width += 1

        queue = new_queue

    return None
```

```python
def benchmark_beam(G, start, goal, beam_width=3):
    if start not in G.nodes or goal not in G.nodes:
        return None, "Start or goal is invalid."

    start_time = time.perf_counter()
    path = beam_search(G, start, goal, beam_width)
    end_time = time.perf_counter()

    if path:
        path_length = len(path) - 1
        run_time = end_time - start_time
        return {
            "Path Length": path_length,
            "Execution Time (s)": run_time,
            "Path": path
        }, None
    else:
        return None, "No path found."

# Run it
result_beam, error_beam = benchmark_beam(G, start, goal, beam_width=3)

if result_beam:
    print("✅ Beam Search Benchmark Results:")
    for k, v in result_beam.items():
        if k != "Path":
            print(f"{k}: {v}")
else:
    print(f"⚠️ {error_beam}")
```

```python
def hill_climbing_search(G, start, goal):
    current = start
    path = [current]
    visited = set([current])

    while current != goal:
        neighbors = [n for n in G.neighbors(current) if n not in visited]
        if not neighbors:
            return None  # stuck

        # Pick neighbor with smallest heuristic
        current = min(neighbors, key=lambda n: manhattan_distance(n, goal))
        path.append(current)
        visited.add(current)

    return path
```

```python
def benchmark_hill_climbing(G, start, goal):
    if start not in G.nodes or goal not in G.nodes:
        return None, "Start or goal is invalid."

    start_time = time.perf_counter()
    path = hill_climbing_search(G, start, goal)
    end_time = time.perf_counter()

    if path:
        path_length = len(path) - 1
        run_time = end_time - start_time
        return {
            "Path Length": path_length,
            "Execution Time (s)": run_time,
            "Path": path
        }, None
    else:
        return None, "No path found (got stuck)."

# Run it
result_hill, error_hill = benchmark_hill_climbing(G, start, goal)

if result_hill:
    print("✅ Hill Climbing Benchmark Results:")
    for k, v in result_hill.items():
        if k != "Path":
            print(f"{k}: {v}")
else:
    print(f"⚠️ {error_hill}")
```
